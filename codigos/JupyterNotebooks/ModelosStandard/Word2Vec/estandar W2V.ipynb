### Funciones para limpiar textos

```python
def docs_to_lists(docs):
    """
    Recibe una lista de strings y regresa
    para cada string una lista limpia
    sin stopwords o puntuaciones para usar
    entrenar modelo.
    """
    temp_nlp = spacy.load("es_core_news_lg")
    doc_corp = []
    for doc in docs:
        clean_doc = []
        for word in temp_nlp(doc):
            if not(word.is_stop or word.is_punct or len(word) == 1):
                clean_doc.append(word.text)
        print(clean_doc)
        doc_corp.append(clean_doc)
    return doc_corp

def doc_to_str(clean_docs):
    """
    Recibe una lista de listas (de la func
    docs_to_lists) y las convierte en un string.
    """
    str_docs = []
    for doc in clean_docs:
        str_docs.append(" ".join(doc))
    return str_docs

def clean_docs(docs):
    """
    Recibe una lista de strings y las limpia de
    puntuaciones y stopwords
    """
    return doc_to_str(docs_to_lists(docs))
```

### Obtenemos todas las palabras para cada pregunta

```python
def get_words(respuestas):
    
    """
    Recibe una lista de respuestas y
    regresa un lista con todas las
    palabras que contiene la lista
    """
    
    palabras = []
    for respuesta in respuestas:
        for palabra in respuesta:
            if palabra not in palabras:
                palabras.append(palabra)
    return palabras
```

```python
palabras = [get_words(respuestas) for respuestas in texts]
```

###  Obtenemos las palabras similares para cada palabra

```python
def get_similar(word, num):
    """
    word es la palabra que vamos a comparar y
    num es la cantidad de palabras que queremos
    obtener
    """
    
    words = word_vectors.most_similar(positive = [f"{word}"], topn = num)
    words = [word[0] for word in words]
    return words

def dict_syn(palabras):
    dictionary = {word : get_similar(word, 10) for word in palabras if word in word_vectors.vocab}
    return dictionary
```

```python
similares = [dict_syn(words) for words in palabras]
```

### Creamos corpi

```python
import pickle
with open("sinonimos.txt", "rb") as fp:
    similares = pickle.load(fp)
similares
```

```python
def make_doc(x):
    """
    Genera un doc para corpus a partir
    del diccionario de siminares (ver 
    resultado abajo)
    """
    new_doc = []
    for key, ls in x.items():
        temp = combinar(key,ls)
        for item in temp:
            new_doc.append(item)
    return new_doc
        
def combinar(word, others):
    syn = 10 ######### syn debe ser el número de sinónimos
    i = syn - 1
    out = []
    while i > 0:
        temp = []
        temp.append(word)
        for n in range(0,i+1):
            temp.append(others[n])
        out.append(temp)
        i -= 1
    
    for i in range(0,syn):
        temp = []
        temp.append(word)
        temp.append(others[i])
        out.append(temp)
    
    return out
```

```python
corpi_syn = [make_doc(diccionario) for diccionario in similares]
```

### Juntamos los corpi con las respuestas

```python
def join_corpi(similares, respuestas):
    doc = []
    for i in similares:
        doc.append(i)
    for i in respuestas:
        doc.append(i)
    return doc
```

```python
with open("respuestas_3.txt", "rb") as fp:
    texts = pickle.load(fp)
texts
```

```python
corpi = [join_corpi(corpi_syn[i],texts[i]) for i in list(range(0,len(texts)))]
corpi
```

### Fucion para calcular la similitudes similarity, wmdistance, n_similarity y distance

```python
def similarity(str1, str2, model):
    """
    str1 debe ser una lista de las palabras
    str2 es la frase modelo
    regresa una lista de listas con las similitudes
    similarity(), wmdistance(), n_sim(), distance()
    en ese orden
    """
    #a = str1.split()
    a = [word for word in str1 if word in model.wv.vocab]
    b = [word for word in str2 if word in model.wv.vocab]
    
    if not len(a) == 0:
        sims = []
        for word in a:
            temp = []
            for word2 in b:
                temp.append(model.similarity(word, word2))
            sims.append(max(temp))
        sim = mean(sims)
        distances = []
        for word in a:
            temp = []
            for word2 in b:
                temp.append(model.wv.distance(word, word2))
            distances.append(max(temp))
        distance = mean(distances)
        wmdistance = model.wmdistance(a,b)
        n_sim = model.n_similarity(a,b)
    else:
        return (0, 0, 0, 0)
    return sim, wmdistance, n_sim, distance

def list_similarity(ls, frases, modelo):
    """
    ls es la lista con todas las respuestas,
    frases son las 2 frases modelos para comparar,
    odelo es el modelo correspondiente al set de respuestas.
    Regresa una lista con la mayor puntuación respecto entre
    las dos frases modelo para cada uno de los 4 algorimtmos de
    similitude.
    """
    
    sim_model1 = [similarity(i, frases[0], modelo) for i in ls]
    sim_model2 = [similarity(i, frases[1], modelo) for i in ls]
    similitudes = []
    for i in range(0,len(sim_model1)):
        sims_respuesta = []
        for n in range(0,len(sim_model1[i])):
            sims_respuesta.append(max(sim_model1[i][n], sim_model2[i][n]))
        similitudes.append(sims_respuesta)
    return similitudes
```
