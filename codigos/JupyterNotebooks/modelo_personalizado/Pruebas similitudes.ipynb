#### Funcion para obtener las palabras similares

```python
def get_similar(word, num):
    """
    word es la palabra que vamos a comparar y
    num es la cantidad de palabras que queremos
    obtener
    """
    
    words = word_vectors.most_similar(positive = [f"{word}"], topn = num)
    words = [word[0] for word in words]
    return words
```

#### Creamos un diccionario con cada palabra y una lista de sus similares.

```python
similares = {word : get_similar(word, 5) for word in cmodel_word if word in word_vectors.vocab}
pprint(similares)
```

```python
def make_doc(x):
    """
    Genera un doc para corpus a partir
    del diccionario de siminares (ver 
    resultado abajo)
    """
    new_doc = []
    for key, ls in x.items():
        temp = combinar(key,ls)
        for item in temp:
            new_doc.append(item)
    return new_doc
        
def combinar(word, others):
    i = 4
    out = []
    while i > 0:
        temp = []
        temp.append(word)
        for n in range(0,i+1):
            temp.append(others[n])
        out.append(temp)
        i -= 1
    
    for i in range(0,5):
        temp = []
        temp.append(word)
        temp.append(others[i])
        out.append(temp)
    
    return out
```
